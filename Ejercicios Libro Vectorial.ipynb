```python
import sympy as smp
import numpy as np
from scipy.optimize import minimize
x,y,z=smp.symbols('x y z')

#Defining our objective function
f=x**2+2*y**2
#Defining our constraint
g=x**2+y**2-1

lamda=smp.symbols('lambda')
L=f-lamda*g
gradL=[smp.diff(L,var) for var in [x,y]]
eq=gradL+[g]

solution=smp.solve(eq,[x,y,lamda],dict=True)
for sol in solution:
    sol['function value']=f.subs({x:sol[x],y:sol[y]})
    sol.pop(lamda)
print(solution)


```
